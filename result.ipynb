```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

means_alt = []
error_alt = []
means_biobj = []
error_biobj = []
means_opt = []
error_opt = []
indexes = []

pathCWD = os.getcwd() + "/data/"
for i in range(2,9):
    indexes.append(i**2)

    # ---- Alternative Biobjective ----
    pathFile = pathCWD + "results/alt_biobj_PacketDelivery_"+ str(i**2) +"x50.dat"
    dataPkt = pd.read_csv(pathFile, sep=",")
    dfPkt = pd.DataFrame(dataPkt)

    delivery = []
    for row in dfPkt.iterrows():
        delivery.append(row[1]['received'] / row[1]['sent'])
    
    means_alt.append(np.mean(delivery) * 100)
    error_alt.append(np.std(delivery) * 100/np.sqrt(len(delivery)))

    # ---- Biobjective ----
    pathFile = pathCWD + "results/biobj_PacketDelivery_"+ str(i**2) +"x50.dat"
    dataPkt = pd.read_csv(pathFile, sep=",")
    dfPkt = pd.DataFrame(dataPkt)

    delivery = []
    for row in dfPkt.iterrows():
        delivery.append(row[1]['received'] / row[1]['sent'])
    
    means_biobj.append(np.mean(delivery) * 100)
    error_biobj.append(np.std(delivery) * 100/np.sqrt(len(delivery)))

    # ---- Optimal ----
    pathFile = pathCWD + "results/opt_PacketDelivery_"+ str(i**2) +"x50.dat"
    dataPkt = pd.read_csv(pathFile, sep=",")
    dfPkt = pd.DataFrame(dataPkt)

    delivery = []
    for row in dfPkt.iterrows():
        delivery.append(row[1]['received'] / row[1]['sent'])
    
    means_opt.append(np.mean(delivery) * 100)
    error_opt.append(np.std(delivery) * 100/np.sqrt(len(delivery)))

fig = plt.figure(figsize=(7, 3))

color1 = '#6B2D97'
color2 = '#96753C'
color3 = '#1E9650'
alpha = 0.9


plt.errorbar(range(len(means_alt)), means_alt, error_alt, color=color1, marker='o', 
             label='Alternative Biobjective', capsize=6, lw=2, capthick=2)

plt.errorbar(range(len(means_alt)), means_biobj, error_biobj, color=color2, marker='o', 
             label='Biobjective', capsize=6, lw=2, capthick=2)

plt.errorbar(range(len(means_alt)), means_opt, error_opt, color=color3, marker='o', 
             label='Minimize UAVs', capsize=6, lw=2, capthick=2)


plt.xticks(range(len(means_alt)), indexes)
plt.xlabel("Virtual Positions")
plt.ticklabel_format()
plt.ylabel("Packet Delivery (%)")
plt.grid(True, 'major', 'y', linestyle='--')
plt.legend(loc='best')

# fig.suptitle('Solving Time by Number of Virtual Positions')
plt.show()
```
