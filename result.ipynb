```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

means_alt = []
error_alt = []
means_biobj = []
error_biobj = []
means_opt = []
error_opt = []
indexes = []

pathCWD = os.getcwd() + "/data/"
for i in range(2,9):
    indexes.append(i**2)

    # ---- Alternative Biobjective ----
    pathFile = pathCWD + "results/alt_biobj_PacketDelivery_"+ str(i**2) +"x50.dat"
    dataPkt = pd.read_csv(pathFile, sep=",")
    dfPkt = pd.DataFrame(dataPkt)

    delivery = []
    for row in dfPkt.iterrows():
        delivery.append(row[1]['received'] / row[1]['sent'])
    
    means_alt.append(np.mean(delivery) * 100)
    error_alt.append(np.std(delivery) * 100/np.sqrt(len(delivery)))

    # ---- Biobjective ----
    pathFile = pathCWD + "results/biobj_PacketDelivery_"+ str(i**2) +"x50.dat"
    dataPkt = pd.read_csv(pathFile, sep=",")
    dfPkt = pd.DataFrame(dataPkt)

    delivery = []
    for row in dfPkt.iterrows():
        delivery.append(row[1]['received'] / row[1]['sent'])
    
    means_biobj.append(np.mean(delivery) * 100)
    error_biobj.append(np.std(delivery) * 100/np.sqrt(len(delivery)))

    # ---- Optimal ----
    pathFile = pathCWD + "results/opt_PacketDelivery_"+ str(i**2) +"x50.dat"
    dataPkt = pd.read_csv(pathFile, sep=",")
    dfPkt = pd.DataFrame(dataPkt)

    delivery = []
    for row in dfPkt.iterrows():
        delivery.append(row[1]['received'] / row[1]['sent'])
    
    means_opt.append(np.mean(delivery) * 100)
    error_opt.append(np.std(delivery) * 100/np.sqrt(len(delivery)))

fig = plt.figure(figsize=(7, 3))

color1 = '#6B2D97'
color2 = '#96753C'
color3 = '#1E9650'
alpha = 0.9


plt.errorbar(range(len(means_alt)), means_alt, error_alt, color=color1, marker='o', 
             label='Alternative Biobjective', capsize=6, lw=2, capthick=2)

plt.errorbar(range(len(means_alt)), means_biobj, error_biobj, color=color2, marker='o', 
             label='Biobjective', capsize=6, lw=2, capthick=2)

plt.errorbar(range(len(means_alt)), means_opt, error_opt, color=color3, marker='o', 
             label='Minimize UAVs', capsize=6, lw=2, capthick=2)


plt.xticks(range(len(means_alt)), indexes)
plt.xlabel("Virtual Positions")
plt.ticklabel_format()
plt.ylabel("Packet Delivery (%)")
plt.grid(True, 'major', 'y', linestyle='--')
plt.legend(loc='best')

# fig.suptitle('Solving Time by Number of Virtual Positions')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

means_alt = []
error_alt = []
means_biobj = []
error_biobj = []
indexes = []
solve_rate = {4: 0.0, 9: 0.0, 16: 0.0, 25: 0.0, 36: 0.0, 49: 0.0, 64: 0.0}
solved_seeds = {4: [], 9: [], 16: [], 25: [], 36: [], 49: [], 64: []}

pathCWD = os.getcwd() + "/data/"
pathFile = pathCWD + "model/output/beta_biobj_solutions.dat"
dataSolAlt = pd.read_csv(pathFile, sep=",")
dfSolAlt = pd.DataFrame(dataSolAlt)

for row in dfSolAlt.iterrows():
    pos = row[1]['numVirtualPositions']
    if row[1]['solveTime'] < 1800.0:
        solve_rate[pos] += 10
        solved_seeds[pos].append(row[1]['seed'])

for i in range(2,9):
    indexes.append(i**2)

    # ---- Biobjective Beta ----
    pathFile = pathCWD + "results/beta_biobj_PacketDelivery_"+ str(i**2) +"x50.dat"
    dataPkt = pd.read_csv(pathFile, sep=",")
    dfPkt = pd.DataFrame(dataPkt)

    delivery = []
    for row in dfPkt.iterrows():
        if row[1]['seed'] not in solved_seeds[i**2]:
            continue
        delivery.append(row[1]['received'] / row[1]['sent'])
    
    means_alt.append(np.mean(delivery) * 100)
    error_alt.append(np.std(delivery) * 100/np.sqrt(len(delivery)))

    # ---- Biobjective Alpha ----
    pathFile = pathCWD + "results/biobj_PacketDelivery_"+ str(i**2) +"x50.dat"
    dataPkt = pd.read_csv(pathFile, sep=",")
    dfPkt = pd.DataFrame(dataPkt)

    delivery = []
    for row in dfPkt.iterrows():
        delivery.append(row[1]['received'] / row[1]['sent'])
    
    means_biobj.append(np.mean(delivery) * 100)
    error_biobj.append(np.std(delivery) * 100/np.sqrt(len(delivery)))


fig, ax1 = plt.subplots(figsize=(7, 3))

color1 = '#6B2D97'
color2 = '#96753C'
color3 = '#1E9650'
alpha = 0.4

ax1.errorbar(range(len(means_alt)), means_biobj, error_biobj, color=color2, marker='o', 
             label='Biobjective Alpha', capsize=6, lw=2, capthick=2, zorder=5)
ax1.errorbar(range(len(means_alt)), means_alt, error_alt, color=color1, marker='o', 
             label='Biobjective Beta', capsize=6, lw=2, capthick=2, zorder=5)

ax1.set_xticks(range(len(means_alt)), indexes)
ax1.set_xlabel("Virtual Positions")
ax1.set_ylabel("Packet Delivery (%)")
ax1.grid(True, 'major', 'y', linestyle='--', color='darkgrey')
ax1.legend(loc='best')

ax2 = ax1.twinx()

ax2.bar(range(len(means_alt)), solve_rate.values(), width=0.6, color=color1, alpha=alpha, zorder=0)

ax2.set_ylabel('Instance Solve Rate (%)', color=color1)
ax2.tick_params(axis='y', which='major', colors=color1)

ax1.set_zorder(ax2.get_zorder()+1)
ax1.set_frame_on(False)
# fig.suptitle('Solving Time by Number of Virtual Positions')
plt.show()
```
